```
import folium
import pandas as pd
```

```
pakistan_map = folium.Map(location = [30.37,69.34],zoom_start=4.5,tiles='OpenStreetMap')
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
df = pd.read_csv('/content/drive/My Drive/IDS/CitiesLongLati.csv')
df
```

```
t = 0
for c in df['Total cases']:
  t = int(c)+int(t)
```

```
for district,tc,long,lat,dc,ac,rc in zip(list(df['District']),list(df['Total cases']),list(df['Longitudes']),list(df['Latitudes']),list(df['Deaths']),list(df['Active cases']),list(df['Recoveries'])):
    folium.CircleMarker(location = [lat,long],
                       radius = (tc/t)*100,
                       color = 'red',
                       fill = False,
                       fill_color = 'blue').add_to(pakistan_map)
    xyz = '<strong>District:  ' + district + '</strong><br>' + 'Total cases: ' + str(tc) + '</strong><br>' + 'Active cases: ' + str(ac) + '</strong><br>' + 'Deaths: ' + str(dc) + '</strong><br>' + 'Recoveries: ' + str(rc)
    iframe = folium.IFrame(xyz, width=200, height=150)
    popupData = folium.Popup(iframe, max_width=500)
    folium.Marker(location = [lat,long], popup=popupData).add_to(pakistan_map)
    

pakistan_map
```
